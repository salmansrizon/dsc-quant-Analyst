- The box plot indicates several outliers in the 'salesPrice' distribution. We can use Turky's HSD test to quantifying the outlier 

``Tukey's HSD (Honestly Significant Difference) is a post-hoc statistical test used after performing an ANOVA (Analysis of Variance) to determine which specific group means are significantly different from each other ``

```python
q25, q75 = np.percentile(df['SalePrice'], [25, 75])  # Calculating the 25th and 75th percentiles
iqr = q75 - q25  # Calculating the interquartile range (I 

for threshold in [1.5, 0.5]:
    lower_bound = q25 - (threshold * iqr)  # Calculating the lower bound for outliers
    upper_bound = q75 + (threshold * iqr)  # Calculating the upper bound for outliers
    outliers = df[(df['SalePrice'] < lower_bound) | (df['SalePrice'] > upper_bound)]  # Identifying outliers
    print(f"Number of outliers with threshold {threshold}: {len(outliers)}")  # Displaying the number of outliers found
```
